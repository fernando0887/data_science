```python
dados = pd.read_csv('dados.csv', sep = ';')
#imprimindo os primeiros itens apenas
dados.head()
```

```python
#conhecendo o tamanho do arquivo base, no caso 187 linhas e 4 colunas
dados.shape
```

```python
#grafico de dispersão usando as colunas pib e valorempenho
x = dados.PIB
y = dados.VALOREMPENHO
sb.scatterplot(x = x, y = y, hue = dados.MUNICIPIO[:10], legend = False)
print(dados.MUNICIPIO[:10]) 
```
