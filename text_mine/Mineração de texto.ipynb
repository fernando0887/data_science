```python
import ssl
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
#Caso haja erro ao tentar fazer o download nltk
try:
    _create_unverified_https_context = ssl._create_unverified_context

except AttributeError:
    
    pass

else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download('stopwords')
```

```python
#Criando o corpus a partir de um conjunto de arquivos
corpus = PlaintextCorpusReader('Arquivos', '.*', encoding = 'ISO-8859-1')
```

```python
#Lendo os arquivos em disco, mostrando os 100 primeiros
f = corpus.fileids()
f
```

```python
#Acessando todas as palavras dos arquivos carregados no corpus
tx = corpus.raw()
ttx = corpus.words()
ttx
```

```python
#acessando um índice específico 
ttx[200]
```

```python
#Obtendo as stopwords em inglês e português
sw = stopwords.words('english', 'portuguese')
sw
```

```python
#Criando a nuvem de palavras
cores = ListedColormap(['blue','yellow','red','green','black'])
nuvem = WordCloud(background_color = 'white', colormap = cores, stopwords = sw, max_words = 100)
nuvem.generate(tx)
plt.imshow(nuvem)
```

```python
#removendo stopwords
pstp = [p for p in ttx if p not in sw]
len(pstp)
```

```python
#removendo pontuação
psp = [p for p in pstp if p not in string.punctuation]
len(psp)
```

```python
#Frequência dos termos mais comuns encontrados no texto
mc = nltk.FreqDist(psp)
print(mc.most_common(100))
```
