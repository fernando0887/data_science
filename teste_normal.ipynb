```python
from scipy import stats
from scipy.stats import norm, skewnorm # = dados n√£o normais
import matplotlib.pyplot as plt
```

```python
#cria√ßao da variavel com dados em ums distribui√ß√£o normal com a fun√ßao rvs
dados = norm.rvs(size = 1000)
dados
```

```python
#histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
#gera√ßao de grafico para verificar se ha distribui√ßao normal
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
#teste de shapiro , hipotese nula dados est√£o normalmente distribuidos (p valor =  ou acima de 0,05) , alternativa nao estao normalmente distribuidos
stats.shapiro(dados)
```
