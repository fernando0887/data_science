```python
from scipy import stats
from scipy.stats import norm, skewnorm # = dados não normais
import matplotlib.pyplot as plt
```

```python
#criaçao da variavel com dados em ums distribuição normal com a funçao rvs
dados = norm.rvs(size = 1000)
dados
```

```python
#histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
#geraçao de grafico para verificar se ha distribuiçao normal
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
#teste de shapiro , hipotese nula dados estão normalmente distribuidos (p valor =  ou acima de 0,05) , alternativa nao estao normalmente distribuidos
stats.shapiro(dados)
```
